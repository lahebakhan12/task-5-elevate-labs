```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
titanic.info()
titanic.describe()
```

### Initial Observations

- The Titanic dataset contains 891 rows and 15 columns.
- Some columns such as age and deck contain missing values.
- The average passenger age is around 30 years.
- Fare values vary widely, indicating possible outliers.

```python
plt.figure()
titanic['age'].hist(bins=30)
plt.title("Age Distribution of Passengers")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure()
sns.boxplot(x=titanic['fare'])
plt.title("Fare Distribution")
plt.show()
```

```python
plt.figure()
sns.countplot(x='survived', hue='sex', data=titanic)
plt.title("Survival Count by Gender")
plt.show()
```

### Visualization Observations

- Most passengers were between 20 and 40 years old.
- Fare distribution shows significant outliers, indicating a few passengers paid very high fares.
- Female passengers had a higher survival rate compared to male passengers.


```python
plt.figure()
sns.scatterplot(x='age', y='fare', hue='survived', data=titanic)
plt.title("Age vs Fare by Survival")
plt.show()
```

```python
sns.pairplot(titanic[['age', 'fare', 'survived']])
plt.show()
```

```python
plt.figure()
sns.heatmap(titanic.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

### Summary of Findings

- Female passengers had a significantly higher survival rate.
- Passengers who paid higher fares were more likely to survive.
- Most passengers were young adults between 20 and 40 years.
- Fare shows a positive correlation with survival.
- The dataset contains missing values, especially in the age and deck columns.
